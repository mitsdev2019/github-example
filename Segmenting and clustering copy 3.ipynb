Implementing Kmeans Clustering to find out the safest places in Toronto

```python
import matplotlib 
import sklearn
from sklearn.cluster import KMeans
downtown_toronto_grouped = df_crime_count_sorted.groupby(['Crime_Count'])
## Cluster Neighborhoods
## Run k-means to cluster the neighborhood into 5 clusters.

# set number of clusters

kclusters = 5

toronto_grouped_clustering = df_crime_count_sorted.drop('Neighbourhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
toronto_merged = df_crime_count_sorted.head(140)
toronto_merged['Cluster Labels'] = kmeans.labels_

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood


toronto_merged
```
